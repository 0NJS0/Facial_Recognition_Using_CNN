```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

```python
model = load_model("C:/Users/NJS/Desktop/CVPR Final Assignment/best_model.h5")

class_names = ["Aurko", "Masum", "Shamim"]
```

```python
IMG_SIZE = 255  
 
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

 
    resized = cv2.resize(frame, (IMG_SIZE, IMG_SIZE))
    normalized = resized.astype("float32") / 255.0
    input_img = np.expand_dims(normalized, axis=0)


    pred = model.predict(input_img)[0]
    predicted_class_index = np.argmax(pred)
    predicted_class_label = class_names[predicted_class_index]
    confidence = pred[predicted_class_index]


    label_text = f"{predicted_class_label} ({confidence*100:.1f}%)"
    cv2.putText(frame, label_text, (30, frame.shape[0] - 30), cv2.FONT_HERSHEY_SIMPLEX,
                1, (255, 255, 0), 2)


    cv2.imshow("Webcam - 3-Class Prediction", frame)

 
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

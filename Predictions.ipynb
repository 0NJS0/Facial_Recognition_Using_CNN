```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model
```

```python
model = load_model("C:/Users/NJS/Desktop/CVPR Final Assignment/final_model.h5")

class_names = ["Masum", "Aurko", "Shamim"]
```

```python
IMG_SIZE = 255  
 
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break


    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)


    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5)

    for (x, y, w, h) in faces:
        face = frame[y:y+h, x:x+w]


        try:
            resized = cv2.resize(face, (IMG_SIZE, IMG_SIZE))
        except Exception as e:
            print(f"Resize failed: {e}")
            continue

        normalized = resized.astype("float32") / 255.0
        input_img = np.expand_dims(normalized, axis=0)


        pred = model.predict(input_img)[0]
        predicted_class_index = np.argmax(pred)
        confidence = pred[predicted_class_index]

        if confidence >= 0.7:
            predicted_class_label = class_names[predicted_class_index]
            label_text = f"{predicted_class_label} ({confidence*100:.1f}%)"
        else:
            label_text = "Unknown"

        # Draw rectangle and label
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(frame, label_text, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX,
                    0.8, (255, 255, 0), 2)

    # Show the frame
    cv2.imshow("Webcam - Face Recognition", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
